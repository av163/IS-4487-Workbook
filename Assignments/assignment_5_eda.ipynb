## 4. Univariate Analysis

Business framing:  

Hotels care about trends like average stay length, customer mix, and pricing. A good EDA starts with understanding single variables and their distribution.

- Select at least 3 individual variables to explore
- Use plots and summary methods (e.g. info(), describe(), etc)  to describe the distribution (hint: we are only looking at the values of one variable, so think of plots you've used in the past that DON'T compare 2 variables.)
- Focus on what matters from a business standpoint (e.g., pricing, cancellations, guest types)

```python
# Your code for univariate analysis (e.g., plots, value counts) 🔧
print(df['is_canceled'].value_counts())


percent_canceled=df['is_canceled'].value_counts(normalize=True)
print(percent_canceled*100)

sns.countplot(x='is_canceled', data=df)
plt.title('Hotel Cancelations')
plt.xlabel('Canceled "0" = Not Canceled, "1" = Canceled')
plt.ylabel('Count')

```

```python
print(df['adr'].describe())

filtered_adr=df[(df['adr'] > 0) & (df['adr'] < 500)]

plt.hist(filtered_adr['adr'],bins=100)
plt.xlabel('ADR')
plt.ylabel('Count')
plt.title('ADR Distribution(excluding extreme outliers)')
plt.show()
```

```python
print(df['adults'].describe())

percent_adults=df['adults'].value_counts(normalize=True)
print(percent_adults*100)

sns.countplot(x='adults', data=df)
plt.title('Adult Distribution')
plt.xlabel('Adults')
plt.ylabel('Count')
```

## 5. Bivariate Analysis

Business framing:  

Stakeholders often ask: “What drives cancellations?” or “Do longer stays mean higher revenue?” Bivariate analysis helps you uncover those kinds of relationships.

- Choose 2 relevant variable pairs (e.g., `lead_time` vs. `is_canceled`, or `adr` vs. `customer_type`)
- Use scatterplots, grouped bar plots, or boxplots to explore the relationships
- Interpret what these relationships could mean for the hotel business

```python
# Your code to analyze variable relationships (e.g., scatterplots, grouped bars) 🔧

filtered_df=df[(df['adr'] > 0) & (df['adr'] < 500)].copy()


filtered_df['total_stay'] = filtered_df['stays_in_weekend_nights'] + filtered_df['stays_in_week_nights']
sns.scatterplot(x='total_stay', y='adr', data=filtered_df)
plt.title('Total Stay vs. ADR')
plt.xlabel('Total Stay')
plt.ylabel('ADR')
plt.show()
```

```python
cancellation_percentages=df.groupby('hotel')['is_canceled'].value_counts(normalize=True).unstack()
print(cancellation_percentages)

plt.figure(figsize=(8, 5))
sns.countplot(x='hotel', hue='is_canceled', data=df)
plt.title('Cancellation Status by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Count')
plt.legend(title='Is Canceled', labels=['No', 'Yes'])
plt.show()
```
